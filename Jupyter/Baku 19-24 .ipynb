```python
import pandas as pd


laps_19 = pd.read_csv('laps with weather19.csv')
laps_21 = pd.read_csv('laps with weather21.csv')
laps_22 = pd.read_csv('laps with weather22.csv')
laps_23 = pd.read_csv('laps with weather23.csv')
laps_24 = pd.read_csv('laps with weather24.csv')
```

```python
from google.colab import files
laps19_24 = pd.concat([laps_19, laps_21, laps_22, laps_23, laps_24], ignore_index=True)
csv_file_name = 'laps19-24.csv'
```

```python
columns_to_keep = ['Driver', 'DriverNumber', 'LapTime', 'LapNumber', 'Stint', 'TyreLife', 'Compound', 'Team','AirTemp','Humidity','Pressure','Rainfall','TrackTemp','WindDirection','WindSpeed']

final_laps19_24 = laps19_24[columns_to_keep]

final_laps19_24['LapTime'] = pd.to_timedelta(final_laps19_24['LapTime'])

# 2. Format 'LapTime' to 'M:SS:MMM' (minutes:seconds:milliseconds)
final_laps19_24['FormattedLapTime'] = final_laps19_24['LapTime'].apply(
    lambda x: f"{int(x.total_seconds() // 60)}:{int(x.total_seconds() % 60):02}:{int(x.microseconds / 1000):03}"
)

```

```python
final_laps19_24 = final_laps19_24.drop('LapTime', axis=1)

# 2. Move 'FormattedLapTime' to the 3rd position
# Get the current column list and rearrange the columns
columns = final_laps19_24.columns.tolist()

# Insert 'FormattedLapTime' at the 3rd position (index 2)
columns.insert(2, columns.pop(columns.index('FormattedLapTime')))

# 3. Reorder the DataFrame columns
final_laps19_24 = final_laps19_24[columns]

# 4. Display the result
csv_file_name = 'final-laps18-24.csv'
final_laps19_24.to_csv(csv_file_name, index=False)  # index=False to exclude the index column

# Download the CSV file
files.download(csv_file_name)
```
