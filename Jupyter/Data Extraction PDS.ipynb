```python
import matplotlib.pyplot as plt
import fastf1 as fastf1
import fastf1.plotting
from google.colab import files

fastf1.plotting.setup_mpl(mpl_timedelta_support=True, misc_mpl_mods=False,
                          color_scheme='fastf1')

session22 = fastf1.get_session(2024, 'Azerbaijan', 'R')
session22.load()

laps = session22.laps.pick_accurate()
weather = session22.weather_data
```

```python
laps['HoursMinutes'] = laps['Time'].apply(lambda x: f"{int(x.total_seconds() // 3600)}:{int((x.total_seconds() % 3600) // 60):02}")

# Display the modified DataFrame
print(laps[['Time', 'HoursMinutes']])

weather['HoursMinutes'] = weather['Time'].apply(lambda x: f"{int(x.total_seconds() // 3600)}:{int((x.total_seconds() % 3600) // 60):02}")

print(weather[['Time', 'HoursMinutes']])
```

```python
import pandas as pd

laps_with_weather24 = pd.merge(
    laps, weather, on='HoursMinutes', how='left', suffixes=('', '_weather')
)

# 2. Display the merged DataFrame
# print(laps_with_weather21)

csv_file_name = 'laps with weather24.csv'
laps_with_weather24.to_csv(csv_file_name, index=False)  # index=False to exclude the index column

# Download the CSV file
files.download(csv_file_name)

```
